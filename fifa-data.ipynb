```python
import numpy as np
import pandas as pd
```

```python
raw_data = pd.read_csv('players_20_cleaned.csv')
```

```python
raw_data.head()
```

```python
data = raw_data.copy()
data.head()
```

```python
data.info()
```

#### initial removal of unneeded data

```python
# drop unneccessary things such as id's, names and ages
data = data.drop(['sofifa_id', 'short_name', 'long_name', 'age'], axis=1)

#drop goalkeeping stats, evaluating outfield players only
data = data.drop(['goalkeeping_diving', 'goalkeeping_handling', 'goalkeeping_kicking', 'goalkeeping_positioning', 'goalkeeping_reflexes'], axis=1)

#drop 'averaged' stats, prevents double inclusion
data = data.drop(['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic'], axis=1)
```

```python
data.info()
```

#### group players by position

```python
data['team_position'].unique()
```

```python
from collections import Counter
Counter(data['team_position'])
```

```python
data_positions_mod = data.copy()
positions = data_positions_mod['team_position']

for i in range(0, 18278):
    if positions[i] == 'RW' or positions[i] == 'LW' or positions[i] == 'RF' or positions[i] == 'LF' or positions[i] == 'RS' or positions[i] == 'LS' or positions[i] == 'CF' or positions[i] == 'ST':
        positions[i] = 'Attacker'
    elif positions[i] == 'LAM' or positions[i] == 'CAM' or positions[i] == 'RAM' or positions[i] == 'LM' or positions[i] == 'RM' or positions[i] == 'LCM' or positions[i] == 'RCM' or positions[i] == 'CM' or positions[i] == 'LDM' or positions[i] == 'CDM' or positions[i] == 'RDM':
        positions[i] = 'Midfielder'
    elif positions[i] == 'LWB' or positions[i] == 'LB' or positions[i] == 'LCB' or positions[i] == 'CB' or positions[i] == 'RCB' or positions[i] == 'RB' or positions[i] == 'RWB':
        positions[i] = 'Defender'
    else:
        positions[i] = 'unknown'
```

```python
from collections import Counter
Counter(positions)
```

```python
# pd.options.display.max_columns = None
# pd.options.display.max_rows = None

# display(positions)
```

#### outfield players only

```python
data_outfield_players = data_positions_mod.copy()

for i in range(18278):
    if data_outfield_players['team_position'][i] == 'unknown':
        data_outfield_players = data_outfield_players.drop([i], axis = 0)
        
data_outfield_players
```

#### dummy variables

```python
data_with_dv = data_outfield_players.copy()
```

```python
positions_columns = pd.get_dummies(data_with_dv['team_position'], drop_first=True)
positions_columns
```

```python
data_with_dv = data_with_dv.drop(['team_position'], axis=1)
data_with_dv['midfielder'] = positions_columns['Midfielder']
data_with_dv['defender'] = positions_columns['Defender']
data_with_dv
```

```python
data_with_dv.columns.values
```

```python
columns_reordered = ['midfielder', 'defender', 'overall', 'weak_foot', 'skill_moves', 'attacking_crossing',
       'attacking_finishing', 'attacking_heading_accuracy',
       'attacking_short_passing', 'attacking_volleys', 'skill_dribbling',
       'skill_curve', 'skill_fk_accuracy', 'skill_long_passing',
       'skill_ball_control', 'movement_acceleration',
       'movement_sprint_speed', 'movement_agility', 'movement_reactions',
       'movement_balance', 'power_shot_power', 'power_jumping',
       'power_stamina', 'power_strength', 'power_long_shots',
       'mentality_aggression', 'mentality_interceptions',
       'mentality_positioning', 'mentality_vision', 'mentality_penalties',
       'mentality_composure', 'defending_marking',
       'defending_standing_tackle', 'defending_sliding_tackle']
```

```python
df_preprocessed = data_with_dv[columns_reordered]
df_preprocessed
```

```python
df_preprocessed.to_csv('fifa-outfield-player-stats-2.csv', index=False)
```
